```python
# PrÃ©parer la sortie vidÃ©o
output_path = 'video_annotee.mp4'
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(3))
height = int(cap.get(4))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec .mp4
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # DÃ©tection
    results = model(frame)
    annotated_frame = results.render()[0]

    # Ã‰crire la frame annotÃ©e dans la vidÃ©o de sortie
    out.write(annotated_frame)

cap.release()
out.release()
```
